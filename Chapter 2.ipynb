```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['EDUCATION'].values.reshape(-1,1), df['default payment next month'].values,
                                                   test_size=0.2, random_state=24)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
np.mean(y_train)
```

```python
np.mean(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
example_lr = LogisticRegression(C=0.1, solver='liblinear')
```

```python
example_lr.fit(X_train, y_train)
```

```python
y_pred = example_lr.predict(X_test)
```

```python
is_correct = y_pred == y_test
np.mean(is_correct)
```

```python
example_lr.score(X_test, y_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```

```python
P = sum(y_test)
P
```

```python
TP = sum((y_test==1) & (y_pred==1))
TP
```

```python
TPR = TP/P
TPR
```

```python
FN = sum((y_test==1) & (y_pred==0))
FN
```

```python
FNR = FN/P
FNR
```

```python
N = sum(y_test==0)
N
```

```python
TN = sum((y_test==0) & (y_pred==0))
TN
```

```python
FP = sum((y_test==0) & (y_pred==1))
FP
```

```python
TNR = TN/N
FPR = FP/N
print('The true negative rate is {} and the false positive rate is {}'.format(TNR, FPR))
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

```python
y_pred_proba = example_lr.predict_proba(X_test)
y_pred_proba
```

```python
prob_sum = np.sum(y_pred_proba, 1)
prob_sum
```

```python
pos_proba = y_pred_proba[:,1]
pos_proba
```

```python
plt.hist(pos_proba)
```

```python
mpl.rcParams['font.size'] = 12
plt.hist(pos_proba)
plt.xlabel('Predicted probability of positive class for testing data')
plt.ylabel('Number of samples')
```

```python
pos_sample_pos_proba = pos_proba[y_test==1]
neg_sample_pos_proba = pos_proba[y_test==0]
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, pos_proba)
```

```python
plt.plot(fpr, tpr, '*-')
plt.plot([0,1], [0,1], 'r--')
plt.legend(['Logistic regression', 'Random chance'])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC curve')
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['LIMIT_BAL'].values.reshape(-1,1), df['default payment next month'].values,
                                                   test_size=0.2, random_state=24)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred_proba = lr.predict_proba(X_test)
```

```python
metrics.roc_auc_score(y_test, y_pred_proba[:,1])
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_proba[:,1])
plt.plot(fpr, tpr, '*-')
plt.plot([0,1], [0,1], 'r--')
plt.legend(['Logistic regression', 'Random chance'])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC curve for LIMIT_BAL feature')
```

```python
precision, recall, thresh_3 = metrics.precision_recall_curve(y_test, y_pred_proba[:,1])
```

```python
plt.plot(recall, precision, '-x')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Logistic regression for LIMIT_BAL feature')
plt.xlim([0,1])
plt.ylim([0,1])
```

```python
metrics.auc(recall, precision)
```
